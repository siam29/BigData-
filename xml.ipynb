```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("XMLReader") \
    .config("spark.jars.packages", "com.databricks:spark-xml_2.12:0.17.0") \
    .getOrCreate()
```

```python
xml_1="offers.677e3cd9b1f7e3.68303544.xml"
xml_2="offers.677e31a2a29e28.53749537.xml"
xml_3="offers.677e378ebeafb7.16469743.xml"
xml_4="offers.677e315922f981.54903227.xml"


df1 = spark.read.format("xml").option("rowTag","unit").load(xml_1)
df2 = spark.read.format("xml").option("rowTag","unit").load(xml_2)
df3 = spark.read.format("xml").option("rowTag","unit").load(xml_3)
df4 = spark.read.format("xml").option("rowTag","unit").load(xml_4)
```
