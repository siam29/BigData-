```python
import json

# Load the amenity_category.json file into a dictionary
with open("dataset/amenity_category.json", "r") as f:
    amenity_category_dict = json.load(f)

flat_amenity_category_dict = {}
for amenity, catagories in amenity_category_dict.items():
    if isinstance(catagories, list):
        for catagory in catagories:
            flat_amenity_category_dict[amenity] = catagory
    else:
        flat_amenity_category_dict[amenity] = catagories
```

```python
from pyspark.sql import functions as F

# Convert the flat_amenity_category_dict into a Spark map
mapping_expr = F.create_map(
    *[F.lit(key_val) for key_val in sum(flat_amenity_category_dict.items(), ())]
)
```

```python
# Map each amenity to its category and ensure unique categories
final = result_df.withColumn(
    "categories",
    F.array_distinct(  # Ensure only unique categories
        F.transform(
            F.col("amenities"),
            lambda amenity: F.coalesce(F.element_at(mapping_expr, amenity), F.lit("Other"))
        )
    )
)
final.show(5,truncate=False)
```

```python
from pyspark.sql.functions import col, expr, array, split, concat_ws, regexp_replace, initcap

# Explode the amenities array into rows, process each string, and reassemble the array
final = (
    final
    .withColumn("amenities", expr("""
        transform(
            amenities,
            x -> initcap(regexp_replace(x, '_', ' '))
        )
    """))
)

# Display the final DataFrame
final.show(truncate=False)
```

```python
# Drop the existing table if it exists
spark.sql(
    """
    CREATE TABLE IF NOT EXISTS local.siamdb.transformed_table (
        themes STRING,
        amenities ARRAY<STRING>,
        categories ARRAY<STRING>,
        country_code STRING
    )
    USING iceberg
    TBLPROPERTIES (
        'history.expire.max-snapshot-age-ms' = '60000',
        'write.format.default' = 'parquet'
    )
    PARTITIONED BY (country_code)
    """
)

# Write the data to the Iceberg table in append mode
final.write \
    .format("iceberg") \
    .mode("overwrite") \
    .saveAsTable("local.siamdb.transformed_table")

print("-> Data written to Iceberg table successfully.")
```
