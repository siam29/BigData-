```python
from pyspark.sql.functions import col, concat_ws, lower, split, size,initcap

# Flatten propertyAmenities and roomAmenities
flattened_df = df3.select(
    col("propertyId.expedia").alias("expedia_id"),
    initcap(concat_ws(", ", col("popularAmenities"))).alias("themes"),  # Capitalize first letter of each word
    concat_ws(", ", *[
        col(f"propertyAmenities.{col_name}") for col_name in df3.schema["propertyAmenities"].dataType.fieldNames()
    ]).alias("property_amenities"),
    concat_ws(", ", *[
        col(f"roomAmenities.{col_name}") for col_name in df3.schema["roomAmenities"].dataType.fieldNames()
    ]).alias("room_amenities")
)

combined_df = flattened_df.select(
    col("expedia_id"),
    col("themes"),
    lower(concat_ws(", ", col("property_amenities"), col("room_amenities"))).alias("amenities"),
    size(split(lower(concat_ws(", ", col("property_amenities"), col("room_amenities"))), ", ")).alias("amenities_count")  # Count the number of amenities
)
# Add amenities_count column by splitting combined_amenities into an array and calculating its size
# Instead of recreating combined_amenities, use the existing column
result_df = combined_df.withColumn(
    "amenities",
    split(col("amenities"), ",\\s*")  # Split the existing combined_amenities column
)


# Show the result
result_df.show(truncate=False)
```

```python
# Transform the combined_amenities column in place
from pyspark.sql.functions import regexp_replace # Import necessary functions

result_df = result_df.withColumn(
    "amenities",
    F.transform( # Call transform through the 'F' alias
        col("amenities"),  # Apply transformation to each element in the array
        lambda x: regexp_replace(x, "\\s+", "_")  # Replace spaces with underscores
    )
)

# Show the updated DataFrame
result_df.show(truncate=False)
```

```python
import json

# Load the amenity_category.json file into a dictionary
with open("/content/drive/MyDrive/Data-W3/amenities/amenity_category.json", "r") as f:
    amenity_category_dict = json.load(f)

flat_amenity_category_dict = {}
for amenity, catagories in amenity_category_dict.items():
    if isinstance(catagories, list):
        for catagory in catagories:
            flat_amenity_category_dict[amenity] = catagory
    else:
        flat_amenity_category_dict[amenity] = catagories
```

```python
flat_amenity_category_dict
```

```python
from pyspark.sql import functions as F

# Convert the flat_amenity_category_dict into a Spark map
mapping_expr = F.create_map(
    *[F.lit(key_val) for key_val in sum(flat_amenity_category_dict.items(), ())]
)

# Map each amenity to its category and ensure unique categories
final = result_df.withColumn(
    "categories",
    F.array_distinct(  # Ensure only unique categories
        F.transform(
            F.col("amenities"),
            lambda amenity: F.coalesce(F.element_at(mapping_expr, amenity), F.lit("Other"))
        )
    )
)
final.show(5,truncate=False)
```
